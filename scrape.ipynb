```python
base_urls = [
	'https://www.angpamantasan.org/news',
	'https://www.angpamantasan.org/opinion',
	'https://www.angpamantasan.org/features',
	'https://www.angpamantasan.org/iskomposisyon',
	'https://www.angpamantasan.org/filipino',
	'https://www.angpamantasan.org/sports',
	'https://www.angpamantasan.org/about-us',
]
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from urllib.parse import urljoin, urlparse
import time

def click_load_more(driver, load_more_button_class):
    while True:
        try:
            # Check if the "Load More" button is present on the page
            load_more_button = driver.find_element(By.CLASS_NAME, load_more_button_class)
            if load_more_button:
                load_more_button.click()
                time.sleep(3)  # Wait for new content to load
            else:
                print("No 'Load More' button found.")
                break
        except Exception as e:
            print(f"No more 'Load More' buttons to click or error occurred: {e}")
            break

def get_all_links(driver, url, load_more_button_class=None):
    # Navigate to the URL
    driver.get(url)
    time.sleep(2)  # Give some time for the page to load

    # Handle "Load More" button if specified
    if load_more_button_class:
        click_load_more(driver, load_more_button_class)
    
    # Find all the links on the current page
    links = driver.find_elements(By.TAG_NAME, "a")
    
    # Extract the URLs from the 'href' attribute
    urls = [link.get_attribute('href') for link in links if link.get_attribute('href')]

    absolute_urls = set()
    for link_url in urls:
        # Ensure the link is absolute
        absolute_url = urljoin(url, link_url)

        # Only visit the link if it's within the same domain
        if not absolute_url.startswith(url):
            continue

        # Add the absolute URL to the set
        absolute_urls.add(absolute_url)

    return absolute_urls
```

```python
# Set up WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

load_more_button_class = "LoadMore_loadMoreButton__6WGOl"  # Replace with the actual class name of the Load More button
urls = set()

for base_url in base_urls:
	urls.update(get_all_links(driver, base_url, load_more_button_class))

# Close the browser
driver.quit()
```
